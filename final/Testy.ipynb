```julia
using BenchmarkTools
using PyPlot
```

```julia
include("power_functions.jl")
include("jacobi_functions.jl")
include("qr_functions.jl")
include("hessenberg_reduction.jl")
```

### Testy za pomocą makra Benchmark dla symetrycznej macierzy (50, 50)

```julia
A = randn(50,50);
# A = A * diagm(1:50) / A;
A = A * A';
```

```julia
Λ_expected, V_expected = eigen(A);
```

```julia
@benchmark power_method(A, 1000)
```

```julia
@benchmark jacobi_A(A, 1000)
```

```julia
@benchmark jacobi_B(A, 1000)
```

```julia
@benchmark QRwithShifts(A, 1000)
```

```julia
@benchmark shiftQRc(A, 1000)
```

```julia
@benchmark shiftQRc_hessen(A, 1000)
```

```julia
@benchmark QR_eigen(A, 1000)
```

```julia
@benchmark QR_eigen_hessen(A, 1000)
```

### Badanie zbieżności metod

Badanie macierzy symetrycznej

```julia
λ_pm, v_pm, Δλ_pm = power_method_err(A, 5000, convert(Float64, Λ_expected[end]));
```

```julia
Λ_j, ΔΛ_j = jacobi_A_err(A, 5000, Λ_expected);
```

```julia
Λ_j_B, ΔΛ_j_B = jacobi_B_err(A, 5000, Λ_expected);
```

```julia
Λ_q, ΔΛ_q = QR_eigen_err(A, 5000, Λ_expected);
```

```julia
Λ_q_h, ΔΛ_q_h = QR_eigen_hessen_err(A, 5000, Λ_expected);
```

```julia
ΔΛ_q_shift_complex = [];
curr_Λ = [];
for i = 1:100
    curr_Λ = convert(Matrix{Float64}, shiftQRc(A, i))
    push!(ΔΛ_q_shift_complex, abs.(sort(curr_Λ, dims=1) - Λ_expected))
end
```

```julia
ΔΛ_q_shift_complex_hessen = [];
curr_Λ = [];
for i = 1:100
    curr_Λ = convert(Matrix{Float64}, shiftQRc_hessen(A, i))
    push!(ΔΛ_q_shift_complex_hessen, abs.(sort(curr_Λ, dims=1) - Λ_expected))
end
```

```julia
ΔΛ_q_shift = [];
curr_Λ = [];
for i = 1:100
    curr_Λ = QRwithShifts(A, i)
    push!(ΔΛ_q_shift, abs.(sort(curr_Λ) - Λ_expected))
end
```

```julia
ΔΛ_j_plot = []
for i = 1:length(ΔΛ_j)
    push!(ΔΛ_j_plot, mean(ΔΛ_j[i]))
end
```

```julia
ΔΛ_j_B_plot = []
for i = 1:length(ΔΛ_j_B)
    push!(ΔΛ_j_B_plot, mean(ΔΛ_j_B[i]))
end
```

```julia
ΔΛ_q_plot = []
for i = 1:length(ΔΛ_q)
    push!(ΔΛ_q_plot, mean(ΔΛ_q[i]))
end
```

```julia
ΔΛ_q_h_plot = []
for i = 1:length(ΔΛ_q_h)
    push!(ΔΛ_q_plot, mean(ΔΛ_q_h[i]))
end
```

```julia
ΔΛ_q_shift_complex_plot = []
for i = 1:length(ΔΛ_q_shift_complex)
    push!(ΔΛ_q_shift_complex_plot, mean(ΔΛ_q_shift_complex[i]))
end
```

```julia
ΔΛ_q_shift_complex_hessen_plot = []
for i = 1:length(ΔΛ_q_shift_complex_hessen)
    push!(ΔΛ_q_shift_complex_hessen_plot, mean(ΔΛ_q_shift_complex_hessen[i]))
end
```

```julia
ΔΛ_q_shift_plot = []
for i = 1:length(ΔΛ_q_shift)
    push!(ΔΛ_q_shift_plot, mean(ΔΛ_q_shift[i]))
end
```

```julia
semilogy(1:length(Δλ_pm), Δλ_pm, "b.-")
semilogy(1:length(ΔΛ_j_plot), ΔΛ_j_plot, "r.-")
semilogy(1:length(ΔΛ_j_B_plot), ΔΛ_j_B_plot, "c.-")
semilogy(1:length(ΔΛ_q_plot), ΔΛ_q_plot, "g.-")
semilogy(1:length(ΔΛ_q_h_plot), ΔΛ_q_h_plot, "m.-")
legend(["Potęgowa", "Jacobi", "Jacobi 2", "QR", "QR - H"])
xlabel("Liczba iteracji")
ylabel("Błąd wartości własnych")
title("Zbieżność badanych metod dla macierzy symetrycznej")
```

```julia
semilogy(1:length(ΔΛ_q_shift_complex_plot), ΔΛ_q_shift_complex_plot, "g.-")
semilogy(1:length(ΔΛ_q_shift_complex_hessen_plot), ΔΛ_q_shift_complex_hessen_plot, "r.-")
semilogy(1:length(ΔΛ_q_shift_plot), ΔΛ_q_shift_plot, "b.-")
legend(["QR shifts complex", "QR shifts complex - H", "QR shifts - H"])
xlabel("Liczba iteracji")
ylabel("Błąd wartości własnych")
title("Zbieżność badanych metod dla macierzy symetrycznej")
```
